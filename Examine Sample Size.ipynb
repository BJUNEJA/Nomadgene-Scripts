```python
### This exercise shows how standard error of means goes down as we increase sample size

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

sample_sizes = np.array([3,10,50,100,500,1000])
mean_sds = np.empty((6))

for N in sample_sizes:
    #1) Draw 1000 samples of size N and calculate the mean for each
    #of those samples, save those means into a vector m
    m=np.zeros(1000)
    for i in range(len(m)):
        #2) calculate the standard deviation of the distribution represented in the vector m
        # numpy.random.randn(...) returns array of shape N, 
            #filled with random floats sampled from a univariate normal 
            #distribution of mean 0 and variance 1
        m[i] = np.mean(np.random.randn(N))
        #print(m[i])
        #3) save the result(sd of the means at given N) into the vector mean.sds    
        itemindex = np.where(sample_sizes==N)
        mean_sds[itemindex] =  np.std(m)
   

plt.plot(sample_sizes, mean_sds)
plt.xlabel('sample size')
plt.ylabel('SEM')
plt.title('SEM vs sample size')
plt.show()


```
